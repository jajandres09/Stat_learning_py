En lugar de los plots anteriores podemos utilizar un heat map que nos permite obtener informacion de la correlacion entre cada una de las variables  que graficamos anteriormente.

```python
#(c)
plot_columnas = ['crim', 'zn' , 'age' , 'medv', 'tax' , 'dis' , 'lstat']

pd.plotting.scatter_matrix(boston[plot_columnas], figsize=(10,10))
plt.show()
```
