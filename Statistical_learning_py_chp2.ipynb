```python
g = sns.PairGrid(college2, vars = college2.iloc[:,1:11],  hue = 'Private')
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter, s = 3)
g.fig.set_size_inches(12,12)
```

##  Box Plots

```python
sns.boxplot(x = 'Private', y = 'Outstate', data = college2)
```

```python
''' Histograma '''
sns.set_style('whitegrid')
g = sns.FacetGrid(college2, hue = 'Private')
g = g.map(plt.hist, 'Outstate', bins = 20, alpha = 0.7)
```

```python
college2['Elite'] = pd.cut(college2['Top10perc'],[0,50,100], labels= ['No','Yes'])
college2['Elite']
```

```python
college2['Elite'].value_counts()
```

```python
college2.boxplot('Outstate', by= 'Elite', figsize = (12,8))
```

```python
fig, ax = plt.subplots(2,2, figsize = (12,10))
college2['Accept'].plot.hist(alpha = 0.5, color = 'blue', bins = 20, ax=ax[0,0])
college2['Enroll'].plot.hist(alpha = 0.5, color = 'red', bins = 20, ax=ax[0,1])
college2['Outstate'].plot.hist(alpha = 0.5, color = 'green', bins = 20, ax=ax[1,0])
college2['Top10perc'].plot.hist(alpha = 0.5, color = 'cyan', bins = 20)
```
